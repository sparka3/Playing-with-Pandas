```python
import pandas as pd
import os.path
```

```python
#get the data from file into a table
file = 'C:\\Users\\ischool-user\\Downloads\\co2_hawaii.txt'
col_name = ['site_code', 'year', 'month', 'day', 'hour', 'minute', 'second', 'value', 'value_unc', 'nvalue', 'latitude', 'longitude', 'altitude', 'elevation', 'intake_height', 'instrument', 'qcflag']
data1 = pd.read_table(file, comment='#', delim_whitespace=True, names=col_name)
print(data1.head())
```

```python
#function to parse the -999.99 values for further calculations
def clean_data(df):
    if df['qcflag'] == '*..':
        return 0
    else:
        return df['value']
```

```python
#calling the clean_data function
data1['value']=data1.apply(clean_data,axis=1)
```

```python
#forming dictionary for storing yearwise C02 level values
d = {}
for i in data1['year'].unique():
    d[i] = [data1['value'][j].round(2) for j in data1[data1['year']==i].index]
```

```python
#calculating mean C02 levels for each year
mean_dic = {}
for key,val in d.items():
    mean_dic[key] = (sum(val)/len(val)).round(2)
```

```python
#calculating max C02 values for each year
max_dic = {}
for key,val in d.items():
    max_dic[key] = (max(val)).round(2)
```

```python
mean_data = pd.DataFrame(list(mean_dic.items()), columns=['Year', 'Hawaii_Mean_Value'])
max_data  = pd.DataFrame(list(max_dic.items()), columns=['Year', 'Hawaii_Max_Value'])
```

```python
hawaii_df=pd.merge(mean_data,max_data,how='left',on='Year')
hawaii_df
```

```python
hawaii_df['%HawaiiChange'] = ((hawaii_df['Hawaii_Mean_Value'] - (hawaii_df['Hawaii_Mean_Value'].shift(1) - 1))/(hawaii_df['Hawaii_Mean_Value'].shift(1) - 1)).fillna(0)
```
