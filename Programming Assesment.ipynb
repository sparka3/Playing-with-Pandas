```python
import pandas as pd
import os.path
```

```python
#get the data from file into a table
file = 'C:\\Users\\ischool-user\\Downloads\\co2_hawaii.txt'
col_name = ['site_code', 'year', 'month', 'day', 'hour', 'minute', 'second', 'value', 'value_unc', 'nvalue', 'latitude', 'longitude', 'altitude', 'elevation', 'intake_height', 'instrument', 'qcflag']
data1 = pd.read_table(file, comment='#', delim_whitespace=True, names=col_name)
filter_data=pd.DataFrame(data1[data1['qcflag']!='*..']).reset_index()
max_df=pd.DataFrame(filter_data.groupby('year')[['value']].max()).reset_index()
mean_df=pd.DataFrame(filter_data.groupby('year')[['value']].mean().round(2)).reset_index()
hawaii_df=pd.merge(max_df,mean_df,on='year')
hawaii_df.columns=['year','hawaii_max','hawaii_mean']
hawaii_df['%hawaiichange'] = ((hawaii_df['hawaii_mean'] - (hawaii_df['hawaii_mean'].shift(1) - 1))/(hawaii_df['hawaii_mean'].shift(1) - 1)).fillna(0)
hawaii_df
```

```python
file = 'C:\\Users\\ischool-user\\Downloads\\co2_alaska.txt'
col_name = ['site_code', 'year', 'month', 'day', 'hour', 'minute', 'second', 'value', 'value_unc', 'nvalue', 'latitude', 'longitude', 'altitude', 'elevation', 'intake_height', 'instrument', 'qcflag']
data2 = pd.read_table(file, comment='#', delim_whitespace=True, names=col_name)
filter_data2=pd.DataFrame(data2[data2['qcflag']!='*..']).reset_index()
max_df2=pd.DataFrame(filter_data2.groupby('year')[['value']].max()).reset_index()
mean_df2=pd.DataFrame(filter_data2.groupby('year')[['value']].mean().round(2)).reset_index()
alaska_df=pd.merge(max_df2,mean_df2,on='year')
alaska_df.columns=['year','alaska_max','alaska_mean']
alaska_df['%alaskaChange'] = ((alaska_df['alaska_mean'] - (alaska_df['alaska_mean'].shift(1) - 1))/(alaska_df['alaska_mean'].shift(1) - 1)).fillna(0)
alaska_df
```

```python
#Merge Hawaii and Alaska dataframes
combined_df=pd.merge(hawaii_df,alaska_df,how='right',on='year')
combined_df
```

```python
combined_df.set_index('year', inplace=True)
combined_df
```

```python
hawaii_perc_mean=combined_df['%hawaiichange'].mean()
alaska_perc_mean=combined_df['%alaskaChange'].mean()
hawaii_perc_mean
```
